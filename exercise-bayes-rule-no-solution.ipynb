## Exercise - Bayes Rule

Peter is going to buy a surprise egg. Surprise eggs are chocolate eggs, which contain a random toy. Some may contain a rare figurine instead. We denote an egg containing a rare figurine by the variable $A = 1$, the absence of a rare figurine by $A = 0$. Peter has heard rumours, that a test (by shaking an egg) might reveal, if it contains a rare figurine or not. The test result is denoted by $B = 1$ (contains a figurine) and $B = 1$ (contains no figurine). The test however is only 70% reliable: In 70% of cases the egg contains a figurine, the test also gave a positive result. In 70% of cases the egg does not contain a figurine, the test was also negative. Additionally, advertisements of those eggs guarantee a figurine to be in every seventh egg.

**Task:**

If Peter applys the test to an egg and the test is positive, what is the probability that Peter really receives a figurine when he buys the egg?

**Hint:**

We are searching for 
$$P(A=1\mid B=1)$$

```python
# Complete this cell

# (...)

# P(B=1|A=1):
pB1_A1 = 0.70
pA1 = 1/7
pB1 = ((1/7)*0.70) + ((6/7)*0.30) 
pA1_B1 = (pB1_A1 * pA1)/pB1
print(pA1_B1)
```
